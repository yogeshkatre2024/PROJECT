#Imagine you have data from a recent state assembly election in India
Implement
1. CREATE LIST TO STORE PARTY NAMES ,CANDIDATES NAME AND VOTE RECEIVED IN EACH CONSISTENCY.
2. USE OPERATOR TO CALCULATE TOTAL VOTES FOR EACH PARTY.IDENTIFY THE WINNIG PARTY IN EACH CONSISTENCY ,AND DEERMINE THE OVERALL OVERALL ELECTION WINNER.
3. CALCULATE VOTE SHARE PERCENTAGE AND IDENTIFY CONSTITUENCY WITH CLOSE CONTESTS(SMALLL VOTE MARGIN I.E:12%)

```python
import pandas as pd
data= {"constituency":["A","B","C","A","B","C","A","B","C"],
       "party":["partyA","partyB","partyA","partyB","partyC","partyC","partyA","partyB","partyA"],
       "candidate":["candidateX","candidateY","candidateZ","candidateP","candidateQ","candidateR","candidateS","candidateT","candidateU"],
       "votes":[15000,12000,18000,13000,10000,11000,17000,14000,16000]}
df=pd.DataFrame(data)
df


```

```python
#calculate total vote for each party
total_votes=df.groupby("party")["votes"].sum()
total_votes
```

```python
#identify winning party in each constituency
#Using idxmax() to get the index of the maximum votes
winning_candidate_indices = df.groupby('constituency')['votes'].idxmax()
winning_party_constituency = df.loc[winning_candidate_indices, ['constituency', 'party']]

print("Winning Party in each Constituency (using idxmax()):")
print(winning_party_constituency)
```

```python
#determine overall winner
overall_winner=total_votes.idxmax()
print("overall Election winner")
overall_winner
```

```python
#calculaye vote share percentage
total_votes=df["votes"].sum()
df['votes share(%)']=(df["votes"]/total_votes)*100
print("vote_share_percentage",df)
```

```python
#Identify constituence with close contests(eg:margin less than 12%)
def close_contest(x):
  vote_counts=x['votes'].values
  if len(vote_counts)>1:
    vote_counts.sort()
    margin=(vote_counts[-1]-vote_counts[-2])/vote_counts[-1]*100
    return margin < 12.0
  else :
    return False


close_constituencies=df.groupby('constituency').filter(close_contest)['constituency'].unique()
print("\nconstituency with closed contest:",close_constituencies)
```
